```python
import pandas as pd
```

```python
df=pd.read_csv("/kaggle/input/for-dummy-and-le/minihomeprices.csv")
df
```

```python
dum=pd.get_dummies(df.town)
dum
df=pd.concat([df,dum],axis='columns')
df
```

```python
X=df.drop(['town','price'],axis='columns')
print(X)
y=df.price
y
```

```python
from sklearn import linear_model
model=linear_model.LinearRegression()
model.fit(X,y)
```

```python
model.predict([[3000,1,0,0]])
```

```python
Y_pred=model.predict(X)
```

```python
df['Y']=Y_pred
df
```

```python
model.score(X,y)
```
